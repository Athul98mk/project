```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Downloads/dataset_players_1.csv')
```

```python
data.head(10)
```

```python
data.describe()
```

```python
data.info()
```

```python
x=data['NaturalFitness']
plt.figure(figsize=(10,8))
ax=sns.countplot(x,color='#00aaaa')

ax.set_xlabel(xlabel='Fitness of Players',fontsize=16)
ax.set_title(label='Distribution of Natural Fitness of Players',fontsize=20)
plt.show()
```

# most consistent players

```python
data.sort_values('Consistency',ascending=False)[['Name','Age', 'Consistency', 'NationID', 'PositionsDesc','IntGoals']].head(25).style.background_gradient('inferno')
```

```python


plt.figure(figsize = (20, 8))
plt.style.use('fivethirtyeight')
ax = sns.countplot('Stamina', data = data, palette = 'bone')
ax.set_xlabel(xlabel = 'Stamina of players', fontsize = 16)
ax.set_ylabel(ylabel = 'Injury   of Players', fontsize = 16)
ax.set_title(label = 'Comparison of Fitness and Injury proness', fontsize = 20)
plt.show()

```

```python
plt.figure(figsize = (20, 9))
plt.style.use('fivethirtyeight')
ax = sns.countplot('Professional', data = data, palette = 'bone')
ax.set_xlabel(xlabel = 'Professional players', fontsize = 20)
ax.set_ylabel(ylabel = 'Quality Passes', fontsize = 20)
ax.set_title(label = 'Passing accuracy of Professionals', fontsize = 25)
plt.show()
```
